```python
import pandas as pd
import glob
import os

# folder with 3 files
input_dir = r"C:\Users\Admin\Documents\VIETTEL\2-crawl-articles\scraper\output-scraper"
output_file = os.path.join(input_dir, "Toàn bộ thông tin mới.xlsx")

# find every file trong folder
excel_files = glob.glob(os.path.join(input_dir, "*.xlsx"))

# read and merge files
dfs = []
for f in excel_files:
    df = pd.read_excel(f)
    dfs.append(df)

merged_df = pd.concat(dfs, ignore_index=True)

# export to new file
merged_df.to_excel(output_file, index=False)
print(f"Merged {len(excel_files)} files into {output_file}")
```
