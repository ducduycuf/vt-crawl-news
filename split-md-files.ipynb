```python
import re
import os

# File markdown gốc
INPUT_FILE = r"C:\Users\Admin\Documents\VIETTEL\2-crawl-articles\output-gemini\Phân tích tổng hợp.md"

# Thư mục output
OUTPUT_DIR = r"C:\Users\Admin\Documents\VIETTEL\2-crawl-articles\output-gemini\md-files"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Đọc nội dung markdown
with open(INPUT_FILE, "r", encoding="utf-8") as f:
    text = f.read()

# Pattern tìm các mục cấp 2 dạng:
# ## 1. Tên mục
pattern = r"(##\s+\d+\.\s+.*)"

# Tìm vị trí các tiêu đề
matches = list(re.finditer(pattern, text))

sections = []

# Tách từng đoạn giữa 2 tiêu đề
for i, match in enumerate(matches):
    start = match.start()
    end = matches[i + 1].start() if i + 1 < len(matches) else len(text)
    sections.append(text[start:end])

# Ghi ra từng file
for sec in sections:
    # Lấy dòng tiêu đề đầu tiên
    first_line = sec.split("\n")[0].strip()

    # Lấy số mục và tên
    m = re.match(r"##\s+(\d+)\.\s+(.*)", first_line)
    if not m:
        continue

    num = m.group(1).zfill(2)
    name = m.group(2)

    # Làm sạch tên file
    clean_name = (
        name.lower()
            .replace(":", "")
            .replace(" ", "_")
            .replace("/", "_")
            .replace("__", "_")
    )

    filename = f"{num}_{clean_name}.md"
    filepath = os.path.join(OUTPUT_DIR, filename)

    # Lưu file
    with open(filepath, "w", encoding="utf-8") as f:
        f.write(sec.strip() + "\n")

    print("✔ Saved:", filename)

print("\nDone! Extracted")
```
