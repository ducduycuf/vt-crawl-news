```python
import pandas as pd
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from datetime import datetime
import markdown2
import re  # Thêm để xử lý regex cho $text$

# -----------------------
# Đọc Excel
# -----------------------
df = pd.read_excel(r"C:\Users\Admin\Documents\VIETTEL\2-crawl-articles\gemini\output-gemini\best-6.xlsx")

# -----------------------
# Hàm tạo HTML cho một section tin tức
# -----------------------
def create_section_html(df_section):
    html = ""
    items = df_section.head(6)

    for i in range(0, len(items), 3):
        row_items = items.iloc[i:i+3]

        html += '<table width="100%" cellpadding="0" cellspacing="0" border="0"><tr>'

        for _, row in row_items.iterrows():
            ngay = row.get('Ngày', '')
            if isinstance(ngay, datetime):
                ngay = ngay.strftime('%d/%m/%Y')

            html += f'''
            <td width="33%" valign="top" style="background:#ffffff; padding:15px; border:1px solid #ddd;">
                <div style="color:#e91e63; font-size:11px; font-weight:bold; margin-bottom:10px;">
                    {row.get('Nguồn','')}
                </div>

                <a href="{row.get('URL','#')}" 
                   style="color:#1a1a1a; font-size:18px; font-weight:700; text-decoration:none; line-height:1.4;">
                   {row.get('Tiêu đề','')}
                </a>

                <div style="font-size:14px; color:#4a4a4a; margin-top:12px; line-height:1.6;">
                    {row.get('Mô tả','')}
                </div>

                <div style="font-size:13px; font-weight:bold; margin-top:12px;">
                    {ngay}
                </div>
            </td>'''

        html += '</tr></table>'

    return html


# -----------------------
# Lọc data theo từng section
# -----------------------
fpt_vnpt_news = create_section_html(df[df['Nguồn'] == 'FPT & VNPT News'])
fpt_vnpt_strategy = create_section_html(df[df['Nguồn'] == 'FPT & VNPT Strategy'])
local_news = create_section_html(df[df['Nguồn'] == 'Tin trong nước'])
global_news = create_section_html(df[df['Nguồn'] == 'Tin quốc tế'])
smart_home_news = create_section_html(df[df['Nguồn'] == 'Smart Home News'])
smart_home_prod = create_section_html(df[df['Nguồn'] == 'Smart Home Products'])


# -----------------------
# Hàm đọc và format Markdown thành HTML đẹp
# -----------------------
def load_md_as_html(path):
    with open(path, "r", encoding="utf-8") as f:
        html_content = markdown2.markdown(f.read())
    
    # Xử lý các thẻ $text$ thành <code> style
    html_content = re.sub(
        r'\$([^$]+)\$',
        r'<code style="background:#f0f0f0; padding:2px 6px; border-radius:3px; font-size:13px; color:#d63384;">\1</code>',
        html_content
    )
    
    # Style cho các heading levels
    html_content = html_content.replace(
        '<h1>',
        '<h1 style="color:#1a1a1a; font-size:20px; font-weight:700; margin:20px 0 15px; border-bottom:3px solid #f61919; padding-bottom:8px;">'
    )
    html_content = html_content.replace(
        '<h2>',
        '<h2 style="color:#f61919; font-size:18px; font-weight:700; margin:20px 0 12px;">'
    )
    html_content = html_content.replace(
        '<h3>',
        '<h3 style="color:#333; font-size:16px; font-weight:600; margin:15px 0 10px;">'
    )
    
    # Style cho paragraph
    html_content = html_content.replace(
        '<p>',
        '<p style="margin:10px 0; line-height:1.7; color:#4a4a4a; font-size:14px;">'
    )
    
    # Style cho ul/ol - cấp 1
    html_content = html_content.replace(
        '<ul>',
        '<ul style="margin:12px 0; padding-left:25px; line-height:1.7;">'
    )
    html_content = html_content.replace(
        '<ol>',
        '<ol style="margin:12px 0; padding-left:25px; line-height:1.7;">'
    )
    
    # Style cho li
    html_content = html_content.replace(
        '<li>',
        '<li style="margin:8px 0; color:#4a4a4a; font-size:14px;">'
    )
    
    # Style cho strong/bold
    html_content = html_content.replace(
        '<strong>',
        '<strong style="color:#f61919; font-weight:700;">'
    )
    html_content = html_content.replace(
        '<em>',
        '<em style="color:#666; font-style:italic;">'
    )
    
    # Wrap trong table với padding khớp với news boxes
    return f'''
    <table width="100%" cellpadding="0" cellspacing="0" border="0" style="margin-bottom:20px;">
        <tr>
            <td style="background:#ffffff; padding:20px; border:1px solid #ddd; border-left:5px solid #f61919;">
                <div style="font-family: Arial, 'Segoe UI', sans-serif; font-size:14px; color:#4a4a4a;">
                    {html_content}
                </div>
            </td>
        </tr>
    </table>
    '''


# -----------------------
# Đọc 6 file Markdown insights riêng
# -----------------------
INS_FPT_NEWS = load_md_as_html(r"C:\Users\Admin\Documents\VIETTEL\2-crawl-articles\gemini\output-gemini\md-files\fpt_&_vnpt_news.md")
INS_FPT_STRATEGY = load_md_as_html(r"C:\Users\Admin\Documents\VIETTEL\2-crawl-articles\gemini\output-gemini\md-files\fpt_&_vnpt_strategy.md")
INS_LOCAL = load_md_as_html(r"C:\Users\Admin\Documents\VIETTEL\2-crawl-articles\gemini\output-gemini\md-files\tin_trong_nước.md")
INS_GLOBAL = load_md_as_html(r"C:\Users\Admin\Documents\VIETTEL\2-crawl-articles\gemini\output-gemini\md-files\tin_quốc_tế.md")
INS_IOT_NEWS = load_md_as_html(r"C:\Users\Admin\Documents\VIETTEL\2-crawl-articles\gemini\output-gemini\md-files\smart_home_news.md")
INS_IOT_PRODUCTS = load_md_as_html(r"C:\Users\Admin\Documents\VIETTEL\2-crawl-articles\gemini\output-gemini\md-files\smart_home_products.md")


# -----------------------
# Load template HTML
# -----------------------
with open(r"C:\Users\Admin\Documents\VIETTEL\2-crawl-articles\mail-template\template_test.html", "r", encoding="utf-8") as f:
    template = f.read()

# -----------------------
# Replace placeholders
# -----------------------
final_html = (
    template
        .replace('{{FPT_VNPT_NEWS}}', fpt_vnpt_news)
        .replace('{{FPT_VNPT_STRATEGY}}', fpt_vnpt_strategy)
        .replace('{{IOT_NEWS}}', smart_home_news)
        .replace('{{SMART_HOME_PRODUCTS}}', smart_home_prod)
        .replace('{{LOCAL_NEWS}}', local_news)
        .replace('{{GLOBAL_NEWS}}', global_news)
        # --- 6 file insights markdown ---
        .replace('{{INS_FPT_NEWS}}', INS_FPT_NEWS)
        .replace('{{INS_FPT_STRATEGY}}', INS_FPT_STRATEGY)
        .replace('{{INS_IOT_NEWS}}', INS_IOT_NEWS)
        .replace('{{INS_IOT_PRODUCTS}}', INS_IOT_PRODUCTS)
        .replace('{{INS_LOCAL}}', INS_LOCAL)
        .replace('{{INS_GLOBAL}}', INS_GLOBAL)
)

# -----------------------
# Gửi email
# -----------------------
msg = MIMEMultipart("alternative")
msg["Subject"] = "Báo cáo Insights Công nghệ & Thị trường"
msg["From"] = "duymlb178@gmail.com"
msg["To"] = "trankhanhha981@gmail.com"

msg.attach(MIMEText(final_html, "html"))

server = smtplib.SMTP("smtp.gmail.com", 587)
server.starttls()
server.login("duymlb178@gmail.com", "xnng rgjg tlvk cumm")
server.sendmail(msg["From"], msg["To"], msg.as_string())
server.quit()

print("✅ Email đã gửi thành công!")
```
